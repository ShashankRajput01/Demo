## 2.3 - Choose a loss function

### Huber loss

```python
def huber_loss(y_true, y_pred, delta=1.0):
    
    error = y_true - y_pred
    huber_condition = delta * (np.abs(error) <= delta)
    quadratic_loss = 0.5 * (error**2)
    linear_loss = delta * (np.abs(error) - 0.5 * delta)
    
    return np.mean(np.where(np.abs(error) <= delta, quadratic_loss, linear_loss))
```

# 2.6 - Building model with deep learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
```

```python
ts_data = df_150['Sales']
```

```python
# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(ts_data)
plt.title('Rossmann Store Sales Time Series Data')
plt.show()
```

```python
# Transform time series data into supervised learning data
def create_supervised_data(data, lag=1):
    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag + 1)]
    columns.append(df)
    df = pd.concat(columns, axis=1)
    df.fillna(0, inplace=True)
    return df
```

```python
# supervised learning data
lag = 1  
supervised_data = create_supervised_data(ts_data, lag)
```

```python
# Scale the data in the (-1, 1) range
scaler = MinMaxScaler(feature_range=(-1, 1))
supervised_data_scaled = scaler.fit_transform(supervised_data)
```

```python
X = supervised_data_scaled[:, :-1]
y = supervised_data_scaled[:, -1]
```

```python
# Reshape input for LSTM
X = X.reshape(X.shape[0], 1, X.shape[1])
```
